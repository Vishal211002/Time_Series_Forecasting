```python
def get_city(address):
    return address.split(',')[1]
```

```python
def get_state(address):
    return address.split(',')[2].split(' ')[1]
```

```python
df['City'] = df['Purchase Address'].apply(lambda x:f'{get_city(x)} {get_state(x)}')
```

```python
df['City']
```

```python
citywise_sales = df.groupby('City')['Sales'].sum()
```

```python
citywise_sales
```

```python
cities = [city for city, df in df.groupby('City')]
```

```python
plt.bar(cities, citywise_sales)
plt.xlabel('City')
plt.ylabel('Sales')
plt.xticks(rotation='90')
```

```python
df_dup = df[df['Order ID'].duplicated(keep=False)]
```

```python
df_dup
```

```python
df_dup['Grouped'] = df_dup.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
```

```python
df_dup['Grouped']
```

```python
df_dup = df_dup[['Order ID','Grouped']].drop_duplicates()
```

```python
df_dup
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
```

```python
for rows in df_dup['Grouped']:
    rows_split = rows.split(',')
    count.update(Counter(combinations(rows_split,3)))
```

```python
for keys,values in count.most_common(10):
    print(keys,values)
```

```python
product_group = df.groupby('Product')
quantity_group = product_group['Quantity Ordered'].sum()

products = [product for product, df in product_group]

plt.bar(products,quantity_group)
plt.xticks(products,rotation=90)
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
price = df.groupby('Product')['Price Each'].sum()
```

```python
fig,ax1 = plt.subplots(figsize=(20,10))

ax2 = ax1.twinx()
ax1.bar(products,quantity_group,color='r')
ax2.plot(products,price,'b-')

ax1.set_xticklabels(products,rotation=90)
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered')
ax2.set_ylabel('Price')

plt.show()
```
